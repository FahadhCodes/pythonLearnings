### Given a list, print the largest and smallest values.

```python
numbers = []
while True:
    numbers.append(int(input("Enter a number: ")))
    cho = input("Enter \"s\" to stop: ").upper()
    if cho == "S":
        break
MIN=MAX = numbers[0]
for x in numbers:
    MAX = max(x,MAX)
    MIN = min(x,MIN)
print(MAX)
print(MIN)
```

### Input a list and a number → count how many times it appears.

```python
numbers=[]
while True:
    numbers.append(int(input("Enter a number: ")))
    CHO = input("Enter \"s\" to Stop storing: ").upper()
    if CHO == "S" :
        break
n = int(input("Enter the number that you want to find the duplicates: "))
print(len([x for x in numbers if x==n]))
```

### Add and Remove Elements


```python
A = set()
while True:
    cho = input("Add(a) OR Remove(r) Stop(s) :").upper()
    if cho == "A":
        A.add(input("Enter something: "))
        print(A)
    elif cho == "S":
        break
    elif cho == "R":
        r = input("Enter item that you want to remove: ")
        try:
            A.remove(r)
        except KeyError:
            print("Their is no such value to remove")
        except:
            print("Something not problem")
        else:
            print(A)
```

### Convert Two Lists to Dictionary
#### keys = ["name", "age"]
#### values = ["Fahad", 21]
#### Create a dictionary using both lists.

```python
keys = ["name", "age"]
values = ["Fahad", 21]
d2 = dict()
lt = keys + values
i=0
while i<len(lt)/2:
    d2.update({lt[i]:lt[i+2]})
    i+=1
print(d2)
```

### Find Key with Max Value
#### Given: Print the key with the highest score.
#### scores = {"A": 30, "B": 50, "C": 20}

```python
scores = {"A": 30, "B": 50, "C": 20}
range1 = scores.values()
range2 = scores.keys()
MAX = scores["A"]
for x in range1:
    MAX = max(MAX,x) 
for x in range2:
    if scores[x] == MAX:
        print(f"{x}:{MAX}")
```
