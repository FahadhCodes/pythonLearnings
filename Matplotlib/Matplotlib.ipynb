# Slot 1 : Basic

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([1,3,5,7,9])
y1 = np.array([2,4,6,8,10])
y2 = np.array([5,2,5,4,8])
color = np.array([50,20,50,40,80])
plt.plot(x,y1,"*:r",ms = 20,mfc = "b",)
plt.scatter(x,y2, c=color, cmap='viridis',ls = ":",marker = "*")
plt.show()
```

# Slot 2 : Layout - 1

```python
import numpy as np
import matplotlib.pyplot as plt
y = np.array([2,5,3,7,5,9])

plt.plot(y,"D-.", color= "#0096FF",mfc = "#00008B",ms = 5)
font1 = {"family":"sans-serif","color":"#00008B"}
font2 = {"family":"monospace","color":"#00008B"}
plt.title("Energy diagram",loc = "left", fontdict = font1, c = "r")
plt.xlabel("Ionization energy",loc = "right", fontdict = font2)
plt.ylabel("Atom Size",loc = "bottom", fontdict = font2)
plt.grid(c = 'green', linestyle = '--', linewidth = 0.5)
plt.show()
```

# Slot 3 : Layout - 2

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

# 1 - used for customized style for each line
# 2 - random styles for each line
x = 2
if x == 1:
    plt.plot(y1,"D-.c",mfc = "r") 
    plt.plot(y2,"d:m",mfc= "b")
elif x == 2:
    plt.plot(x1, y1, x2, y2)
plt.show()
```

# Slot 4 : Layout - 3 

```python
import numpy as np
import matplotlib.pyplot as plt

y1 = np.array([2,5,3,7,5,9])
y2 = np.array([45,25,35,15,25,10])
y3 = np.array([2,5,3,7,4,9,2])

plt.suptitle("Main Title")

#plot 1
plt.subplot(2,2,1)
plt.plot(y1)
plt.title("y1")

#plot 2
plt.subplot(2,2,2)
plt.plot(y2)
plt.title("y2")

#plot 3
plt.subplot(2,2,3)
plt.plot(y3)
plt.title("y3")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

fig = plt.figure()
fig.suptitle("Controlling subplot sizes with width_ratios and height_ratios")
y1 = np.array([2,5,3,7,5,9])
y2 = np.array([45,25,35,15,25,10])
gs = GridSpec(2, 2, width_ratios=[1, 2], height_ratios=[4, 1])
ax1 = fig.add_subplot(gs[0])
plt.plot(y1)
ax2 = fig.add_subplot(gs[1])
plt.plot(y2)
ax3 = fig.add_subplot(gs[2:])
plt.plot(y1)

plt.show()
```

# Slote 4 : Other Data visualizations

### Scatter Diagram

```python
import numpy as np
from matplotlib import pyplot as plt
x = np.array([1,2,3,4,5,6])
y = np.array([2,5,3,7,5,9])
colors =np.array([20,50,30,70,50,90]) 
sizes = np.array([20,50,30,70,50,90])
plt.scatter(x, y, c=colors, cmap="magma", s = sizes)
plt.colorbar()
plt.show()
```

### Bar charts

```python
x = np.array(["A","B","C","D","E"])
y = np.array([10,50,30,20,5])
plt.subplot(2,1,1)
plt.bar(x,y, width=0.5, color = "m")
plt.subplot(2,1,2)
plt.barh(x,y, height=0.8, color = "g")
plt.show()
```

```python
x1 = np.random.normal(170, 10, 250)
plt.hist(x1,color = "y")
plt.show()
```

### Pie Chart

```python
x = np.array([12,34,80,40])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["g","y","r","k"]
exp = [0.1,0,0.025,0]
plt.pie(x, labels = mylabels,colors = mycolors, explode = exp, shadow = True)
plt.legend(title = "Fruits")
plt.show()
```
